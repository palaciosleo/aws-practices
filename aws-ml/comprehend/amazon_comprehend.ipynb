```python
import boto3
import pandas as pd
```

```python
## Selecting profile credentials
session = boto3.session.Session(profile_name='aws_academy')
```

```python
## Selecting AWS Service
comprehend = session.client(service_name='comprehend')
```

## Detecting the Dominant Language

```python
domlan_1 = 'Esto es una test just to see if AWS puede detectar cual es el lenguaje dominante'
domlan_2 = 'This is a test para ver si AWS can detect which is the dominant languaje'
```

```python
response_1 = comprehend.detect_dominant_language(Text = domlan_1)
```

```python
response_1
```

```python
[(x['LanguageCode'], x['Score']) for x in response_1['Languages']]
```

```python
response_2 = comprehend.detect_dominant_language(Text = domlan_2)
```

## Detecting Named Entities

```python
namen = """Hola, mi nombre es Leonardo Palacios y tengo 33 años. 
Vivo en Córdoba, Argentina, trabajo como Data Scientist en Ithreex Global.
Mi número de teléfono es 0351-5968741 y me cuenta bancaria es xxxxx9879879787"""
```

```python
entities_response = comprehend.detect_entities(Text=namen, LanguageCode='es')
```

```python
entities_response
```

```python
for entidad in [( x['Type'], x['Text'], x['Score']) for x in entities_response['Entities']]:
    print("Tipo: ", entidad[0], '|', 'Texto:', entidad[1], '|', 'Confianza:', entidad[2])
```

## Detecting Personally Identifiable Information (PII)

```python
pii_response = comprehend.detect_pii_entities(Text=namen, LanguageCode='en') # Por algun motivo no reconoce 'es'
```

```python
pii_response
```

```python
for pii in [(x['Type'], x['Score'], x['BeginOffset'], x['EndOffset']) for x in pii_response['Entities']]:
    print("Tipo: ", pii[0], '|', 'Confianza:', pii[1], '|', 'Texto:', namen[pii[2]:pii[3]])
```

## Determine Sentiment

Single-Document Processing – You call Amazon Comprehend with a single document and receive a synchronous response.

Multiple Document Synchronous Processing – You call Amazon Comprehend with a collection of up to 25 documents and receive a synchronous response.

Asynchronous Batch Processing – You put a collection of documents into an Amazon S3 bucket and start an asynchronous operation to analyze the documents. The results of the analysis are returned in an S3 bucket.
